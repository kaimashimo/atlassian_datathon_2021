```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('WorldCupMatches.csv')

df.head()
```

```python
def winner(row):
    if row[6] > row[7]:
        ret = row[18]
    elif row[6] == row[7]:
        ret = None
    else:
        ret = row[19]
    return ret
```

```python
# df.loc()[0][19]
```

```python
df['Winner'] = df.apply (lambda row: winner(row), axis=1)
```

```python
participation = pd.DataFrame({c:[] for c in np.append(np.array(['Year']), df['Home Team Name'].unique())})
participation['Year'] = []

for year in df.loc[df.Year>=1960]['Year'].unique():
    countries = df[df['Year']==year]['Home Team Name'].unique()
    participation = participation.append(dict({'Year': year},**{country:(1 if (country in countries) else 0) for country in countries}), ignore_index=True)

participation = participation.fillna(0)
```

```python
participation['ONE'] = 1
```

```python
participation
```

```python
cum_part = participation.groupby('ONE').cumsum()
cum_part['Year'] = participation['Year']
```

```python
cum_part
```

```python
cum_part.plot(x='Year',figsize=(20,20))
```
