```python
import pandas as pd
import numpy as np
import os
```

```python
big = pd.DataFrame([])
for country in os.listdir('team-stats-new/team-stats/')[1:]:
    print(country)
    new = pd.read_csv('team-stats-new/team-stats/'+country)
    new['fifa_team'] = country[:-4]
    big = big.append(new)
```

```python
big = big[big['Overall'].notna()]
big = big[big['Release Clause'].notna()]
```

```python
big.columns
```

```python
df = big
```

```python
def money_string_to_num(i):
    return float(str(i)[1:-1])/1000 if str(i)[-1]=='K' else float(str(i)[1:-1])
```

```python
df['Release Clause'] = df['Release Clause'].apply(money_string_to_num)
```

```python
df[df['fifa_team']=='new-zealand']
```

```python
intermediate = pd.concat([df.groupby('fifa_team').median(),df.groupby('fifa_team').size()],axis=1)
```

```python
# agg = intermediate[['Overall','Release Clause',0]].loc[intermediate[['Overall','Release Clause',0]][0] > 10]
agg = intermediate[['Overall', 'Release Clause']]
```

```python
regions = pd.read_csv('all-teams.csv')[['Country','Region']]
regions['Country']=regions['Country'].str.lower()
regions.Region = regions['Region'].str.replace('UFA','UEFA')
regions.Country = regions.Country.str.replace(' ','-')
```

```python
regions
```

```python
j = agg.merge(regions,how='left',left_on='fifa_team',right_on='Country')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(np.array(j['Release Clause']).reshape(-1, 1))

j['Release Clause'] = scaler.transform(np.array(j['Release Clause']).reshape(-1, 1))

scaler = StandardScaler()
scaler.fit(np.array(j['Overall']).reshape(-1, 1))

j['Overall'] = scaler.transform(np.array(j['Overall']).reshape(-1, 1))
```

```python
j['value'] = j['Overall']+j['Release Clause']
```

```python
j = j.rename(columns={'Country':'country', 'Region':'region', 'value':'value', 0:'count'})
j = j[['country','region','value']]
```
